```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from keras.models import load_model
```

```python
model = load_model('model.h5')
model.summary()
```

```python
img_name = "Harrison_Ford_0007"

img = Image.open("test_imgs/" + img_name + ".jpg").resize((224,224))
plt.figure(figsize = (15, 4))
plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Input")

pred = model.predict(np.expand_dims(np.asarray(img), axis=0))
plt.subplot(1,3,2)
plt.imshow(pred.reshape(224,224) > .9, cmap='gray')
plt.title("Prediction")

label = np.asarray(Image.open("test_gt/" + img_name + ".ppm").resize((224,224)))[:,:,1]
plt.subplot(1,3,3)
plt.imshow(label, cmap='gray')
plt.title("Ground Truth")
plt.show()
```
